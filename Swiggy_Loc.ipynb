```python
sort_df =bellandur_df.groupby(['Restaurant Name','Restaurant Id','Category Name','Area','Month'], as_index = False)['No. of Orders'].count().sort_values(by="No. of Orders", ascending = False)
```

```python
sort_df_filter = sort_df['No. of Orders'] > 10
```

```python
sort_df_Bel=sort_df[sort_df_filter]
```

## Number of Sandwiches Sold as compared to other Categories in Bellandur

```python
sandwiches = ["Grilled Sandwiches(3 Slice Bread)","Grilled Sandwiches(2 Slice Bread)","Hearty Sandwiches"]
sort_df_Bel.columns
sort_df_Bel['Category Name']=sort_df_Bel['Category Name'].replace(to_replace=sandwiches,value = 'Sandwiches')

df_cat=sort_df_Bel.groupby(['Category Name',"Area"], as_index = False)['No. of Orders'].sum().sort_values(by='No. of Orders',ascending = False)
plot_df=df_cat.head(20)

#filter_sandwich = df_cat['Category Name'].str.contains('Sandwiches')
#df_cat[filter_sandwich]

fig,ax = plt.subplots()
plt.figure(figsize=(30,10))
plt.title("Top 20 Product Categories Popular in Bellandur")
plt.ylabel("Count of Orders for Category in the Span of 3 months")
plt.xlabel("Category Name")
plt.grid()



plt.bar(plot_df['Category Name'].values, plot_df['No. of Orders'].values, color = ['g','b','r','orange','cyan','black'])
fig = plt.gcf()
ax = fig.gca()
ax.annotate(302,xy=("Sandwiches",302),xycoords = "data", xytext = ("Sandwiches", 1000), textcoords ="data",
           arrowprops = dict(arrowstyle = "->", connectionstyle = "arc3"))
plt.show()
```
